```python
X = data.loc[:, ['Temperature','Density','Absolute Viscosity', 'Kinematic Viscosity','Prandtl Number','Specific Heat','Thermal Conductivity','Orifice Velocity','Velocity of Air','Reynolds Number','Nusselt Number','h','m']]
y = data[['Effectiveness']]
```

```python
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.6, test_size = 0.4, random_state = 100)

print(X), print(y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print(lm.coef_)
```

```python
coef_df = pd.DataFrame(zip(X_train.columns, lm.coef_))
coef_df.transpose()
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
y_train_pred = lm.predict(X_train)

from sklearn import metrics
train_accuracy = lm.score(X_train, y_train)
print(train_accuracy)
```

```python
y_test_pred = lm.predict(X_test)
test_accuracy = lm.score(X_test, y_test)
print(test_accuracy)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test_pred)
```

```python
Df_y_test_pred = pd.DataFrame(y_test_pred,columns=['Predicted Effectiveness'])
X_test.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
Df_y_test_pred.reset_index(drop=True, inplace=True)
Test_with_pred = pd.concat([X_test, y_test, Df_y_test_pred], axis = 1)
Test_with_pred
```
